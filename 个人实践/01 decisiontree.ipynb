## 剪枝

```python
clf = tree.DecisionTreeClassifier(criterion='entropy'
                                  ,random_state= 30
                                  , splitter= 'random' 
                                  , max_depth= 3
                                  , min_samples_leaf=5
                                  , min_samples_split=10)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)    # max_depth就是最大的深度  min_sample_leaf就是每个子节点至少要5个样本点才分
#min_samples_split 就是每个节点至少要10个样本才会分
score
```

```python
import matplotlib.pyplot as plt
test = []
for i in range(10):
    cls = tree.DecisionTreeClassifier(criterion = 'entropy'
                                      ,max_depth=3 +i
                                      ,random_state= 30
                                      ,min_samples_leaf=5
                                      ,min_samples_split=5
                                      ,splitter='random')
    cls = cls.fit(Xtrain, Ytrain)
    score = cls.score(Xtest, Ytest)
    test.append(score)

plt.plot(range(1,11), test, color = 'red', label = 'max_depth')
plt.legend()
plt.show()
```

对于标签不平衡很多是用class_weight赋予更大的权重 默认的话所有权重都相等
min_weight_fraction_leaf就是给有权重的不同的class进行剪枝的

网格搜索能够帮助我们同时调整多个参数的技术 枚举技术

```python
import numpy as np

#参数变量就是希望网格搜索的参数范围

parameters = {
    'criterion':('entropy','gini')
    ,'splitter':('best', 'random')
    ,'max_depth':[*range(1,10)]   # 等价于list(range(1,10))
    ,'min_samples_leaf':[*range(1,50,5)]
    ,'min_impurity_decrease':[*np.linspace(0,0.5,51)]  # 都是等差数列 linspace最后一个是点数 arange是步长
}

clf = DecisionTreeClassifier()
GS = GridSearchCV(clf, parameters, cv=10)
GS = GS.fit(Xtrain, Ytrain)
```
