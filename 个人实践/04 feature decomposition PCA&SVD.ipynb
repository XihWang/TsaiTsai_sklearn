```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
#  x2  ()
x.shape
```

```python
#  x4
import pandas as pd
pd.DataFrame(x)
```

```python
pca = PCA(2)
pca = pca.fit(x)
X_dr = pca.transform(x)
```

```python
X_dr.shape
```

```python
y
```

```python
X_dr
```

```python
X_dr[y==0, 0]
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
plt.scatter(X_dr[y==0,0], X_dr[y==0,1], c='red', label=iris.target_names[0])
plt.scatter(X_dr[y==1,0], X_dr[y==1,1], c='black', label=iris.target_names[1])
plt.scatter(X_dr[y==2,0], X_dr[y==2,1], c='orange', label=iris.target_names[2])
plt.legend()
plt.title('PCA of IRIS')
plt.show()
```

```python
pca.explained_variance_  # 
```

```python
pca.explained_variance_ratio_  # 
```

```python
pca.explained_variance_ratio_.sum()
```

```python
pca_line = PCA().fit(x)
pca_line.explained_variance_ratio_  # 
```

```python
import numpy as np
pca_line = PCA().fit(x)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4]) #
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
# 
```

```python
rng = np.random.RandomState(42)
noisy = rng.normal(digits.data, 2)  # 2 np.random.normal
figure, axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
for i, ax in enumerate(axes.flat):  # axes410 1
    ax.imshow(noisy[i,:].reshape(8,8), cmap = 'binary')
```

```python
pca = PCA(0.5, svd_solver='full').fit(noisy)
X_dr = pca.transform(noisy)
```

```python
X_dr.shape
```

```python
X_inverse = pca.inverse_transform(X_dr)
```

```python
X_inverse.shape
```
