## PCA

为了让样品方差的无偏估计正好是总体方差

方差越大 所包含的信息就越多

PCA是看方差 SVD是看奇异值

特征压缩是特征创造， 特征选择就是特征选择

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
# 作为数组 x是2维的  (有几个数字就是几维)
x.shape
```

```python
# 作为数据表或特征矩阵 x是4维的
import pandas as pd
pd.DataFrame(x)
```

```python
pca = PCA(2)
pca = pca.fit(x)
X_dr = pca.transform(x)
```

```python
X_dr.shape
```

```python
y
```

```python
X_dr
```

```python
X_dr[y==0, 0]
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
plt.scatter(X_dr[y==0,0], X_dr[y==0,1], c='red', label=iris.target_names[0])
plt.scatter(X_dr[y==1,0], X_dr[y==1,1], c='black', label=iris.target_names[1])
plt.scatter(X_dr[y==2,0], X_dr[y==2,1], c='orange', label=iris.target_names[2])
plt.legend()
plt.title('PCA of IRIS')
plt.show()
```

```python
pca.explained_variance_  # 每个新特征向量上所带的信息
```

```python
pca.explained_variance_ratio_  # 所占原始所有信息百分比
```

```python
pca.explained_variance_ratio_.sum()
```

```python
pca_line = PCA().fit(x)
pca_line.explained_variance_ratio_  # 就是每个特征的重要性
```

```python
import numpy as np
pca_line = PCA().fit(x)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4]) #这是为了限制坐标轴显示为整数
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
# 应该选转折点的降维维度
```

### 最大似然法求超参数n
除了输入整数，n_components还有哪些选择呢？之前我们提到过，矩阵分解的理论发展在业界独树一帜，勤奋智
慧的数学大神Minka, T.P.在麻省理工学院媒体实验室做研究时找出了让PCA用最大似然估计(maximum likelihood
estimation)自选超参数的方法，输入“mle”作为n_components的参数输入，就可以调用这种方法。

```python
pca_mle = PCA(n_components='mle')  # 但计算量会变大
x_mle = pca_mle.fit_transform(x)
x_mle
```

```python
pca_mle.explained_variance_ratio_.sum()
```

```python
pca_f = PCA(n_components=0.97, svd_solver='full')
x_f = pca_f.fit_transform(x)
```

```python
x_f
```

```python
pca_f.explained_variance_ratio_.sum()
```

### 人脸识别的降维与复原

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
faces
```

```python
faces.data
```

```python
faces.data.shape # 行是样本 列是样本相关的所有特征
```

```python
faces.images.shape #1348是矩阵中图像的个数  62是每个图像的特征矩阵的行  47是每个图像的特征矩阵的列
```

```python
62*47  # 所以data的列其实是每一个像素点的值
```

```python
fig, axes = plt.subplots(3,8
                        ,figsize = (8,4) # 行是第一个  列长是第二个
                        ,subplot_kw = {'xticks':[], 'yticks':[]} # 不要显示坐标轴
                       )
for i,ax in enumerate(axes.flat):
    ax.imshow(faces.images[i,:,:], cmap = 'gray')   #没有cmap=‘gray’最后就是绿色的 因为只有一个通道有颜色
```

```python
fig
```

```python
axes
```

```python
axes[0][0].imshow(faces.images[0,:,:])
```

```python
[*axes.flat]
```

```python
pca = PCA(150
         )
```

```python
pca.fit(faces.images)  # pca只能降维2维数组
```

```python
pca.fit(faces.data) 
```

```python
V = pca.components_
```

```python
V.shape  # 乘以原有矩阵就是降维后的矩阵
```

```python
V
```

```python
X_dr = pca.transform(faces.data)
```

```python
fig, axes = plt.subplots(3,8
                        ,figsize = (8,4) # 行是第一个  列长是第二个
                        ,subplot_kw = {'xticks':[], 'yticks':[]} # 不要显示坐标轴
                       )
for i,ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47), cmap = 'gray')   #没有cmap=‘gray’最后就是绿色的 因为只有一个通道有颜色
# 下图是每个特征向量的长什么样（resize到62 47） 可以看出五官和亮度很重要
```

```python
faces_back = pca.inverse_transform(X_dr)
faces_back.shape
```

```python
figure, axes = plt.subplots(2,10,figsize=(10,2.5),subplot_kw={'xticks':[],'yticks':[]})
for i in range(10):
    axes[0,i].imshow(faces.images[i,:,:], cmap='binary_r')
    axes[1,i].imshow(faces_back[i,:].reshape(62,47), cmap = 'gray')

# 明显上面更清晰 因为上面2900特征 下面就150特征重新升维（所以inverse不是完全逆转） 
    #但是仍然能够区分出来是谁 所以人脸识别都是提取少量特征与原先系统录入的大量特征比较 从而加快识别
```

```python
axes
```

```python
for i in range(10):
    axes[0,i].imshow(faces.images[i,:,:])
    axes[1,i].imshow(faces_back[i,:].reshape(62,47))
```

### 用pca降噪

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
digits = load_digits()
digits.data.shape
```

```python
digits.target
```

```python
pd.Series(digits.target).unique()
```

```python
set(digits.target)
```

```python
[1,1,23,4,5,3,3,2,4].unique()  # unique方法是series独有的
```

```python
digits.images.shape
```

```python
list(enumerate(axes))
```

```python
figure, axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
for i, ax in enumerate(axes.flat):  # axes是4行10列的 记得降到1维
    ax.imshow(digits.images[i,:,:], cmap = 'binary')
```

```python
rng = np.random.RandomState(42)
noisy = rng.normal(digits.data, 2)  # 原始数据中加入方差为2的噪音 np.random.normal
figure, axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
for i, ax in enumerate(axes.flat):  # axes是4行10列的 记得降到1维
    ax.imshow(noisy[i,:].reshape(8,8), cmap = 'binary')
```

```python
pca = PCA(0.5, svd_solver='full').fit(noisy)
X_dr = pca.transform(noisy)
```

```python
X_dr.shape
```

```python
X_inverse = pca.inverse_transform(X_dr)
```

```python
X_inverse.shape
```

```python
figure, axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
for i, ax in enumerate(axes.flat):  # axes是4行10列的 记得降到1维
    ax.imshow(X_inverse[i,:].reshape(8,8), cmap = 'binary')  # 达到了一定的降噪目的
```
